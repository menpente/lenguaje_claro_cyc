```python
with open("Intro.txt", mode='r', encoding='utf-8-sig') as file: #Abrimos el archivo en modo lectura (usamos "with" para que se cierre automáticamente al terminar; incluimos: encoding='utf-8-sig' para evitar que incluya los caracteres BOOM u'\ufeff')
    texto = file.read() #Renombramos el output
print(texto) #Visualizamos el texto
```

```python
texto.split(".") #Separamos las frases del texto y creamos una lista en la que cada elemento es una de las frases. 
```

```python
frases = texto.split(".") #Renombramos el output
```

```python
frases.pop() #Eliminamos el último elemento de la lista que se añade por defecto
```

```python
len(frases) #Calculamos el número de frases (items) de la lista
```

```python
palabras = [len(sentence.split()) for sentence in frases] #Calculamos el número de palabras por cada frase (item de la lista)
print(palabras) #Visualizamos el output
```

```python
frases[2] #Podemos realizar búsquedas de las frases que nos interesen en la lista
```

```python
for frase, palabra in zip(frases, palabras): #Obtenemos el par de valores frase-palabras (número de palabras de cada frase)
	print(frase, palabra)
```

```python
for frase, palabra in zip(frases, palabras): #Con este bucle filtramos las frases que tienen una longitud superior a 30 palabras
  if palabra > 30:
	  print(frase, palabra)
```
